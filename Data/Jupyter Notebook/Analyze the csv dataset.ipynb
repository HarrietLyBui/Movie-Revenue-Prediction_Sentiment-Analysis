<h2>Match Naive Bayes Sentiment Score with our final dataset</h2>

```python
nb_data = pd.read_csv("dataset_nb.csv")
nb_data.head()
```

```python
final_data = pd.read_csv('Final_Dataset.csv')
```

```python
final_data = final_data.assign(NB_sent_score = [i for i in final_data['Movie']])
final_data.head()
```

```python
index = nb_data[nb_data['Movie']=='The Bridge on the River Kwai'][['NB_score_normalize']].index
float(nb_data['NB_score_normalize'].iloc[index])
```

```python
type(nb_data['NB_score_normalize'].iloc[index])
```

```python
float(nb_data.iloc[index, -1])
```

```python
for name in final_data['Movie']:
    index = nb_data[nb_data['Movie']==name][['NB_score_normalize']].index
    for score in nb_data['NB_score_normalize'].iloc[index]:
        final_data.loc[final_data['Movie'] ==name, 'NB_sent_score'] = score
```
